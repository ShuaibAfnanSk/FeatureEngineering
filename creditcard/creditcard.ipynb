```python
from collections import Counter
merchant_df = df.groupby('merchant')['is_fraud'].apply(list).reset_index()
merchant_counts = []
for i in range(merchant_df.shape[0]):
    merchant_counts.append(Counter(merchant_df['is_fraud'][i]))
```

```python
new_merchant_df = pd.DataFrame(merchant_counts)
merchant_df = pd.concat([merchant_df,new_merchant_df],axis=1)
merchant_df.drop('is_fraud',inplace=True,axis=1)
merchant_df['ratio'] = ((merchant_df[0] /(merchant_df[0] + merchant_df[1]) * 100))
```

```python
merchant_df.sort_values(by='ratio',ascending=True)
```

```python
category_df = df.groupby('category')['is_fraud'].apply(list).reset_index()
```

```python
new_cat_counts = []
for i in range(len(category_df['is_fraud'])):
    new_cat_counts.append(Counter(category_df['is_fraud'][i]))
```

```python
category_new_df = pd.DataFrame(new_cat_counts)
```

```python
category_df = pd.concat([category_df,category_new_df],axis=1)
```

```python
category_df.drop('is_fraud',inplace=True,axis=1)
```

```python
category_df['ratio'] = ((category_df[0] /(category_df[0] + category_df[1]) * 100))
category_df.sort_values(by='ratio',ascending=False)
```

```python
gender_df = df.groupby('gender')['is_fraud'].apply(list).reset_index()
```

```python
gender_counts = []
for i in range(gender_df.shape[0]):
    gender_counts.append(Counter(gender_df['is_fraud'][i]))
new_gender_df = pd.DataFrame(gender_counts)
gender_df = pd.concat([gender_df,new_gender_df],axis=1)
gender_df.drop(['is_fraud'],axis=1,inplace=True)
gender_df['ratio'] = ((gender_df[0] /(gender_df[0] + gender_df[1]) * 100))
gender_df.sort_values(by='ratio',ascending=True)
```

```python
job_df = df.groupby('job')['is_fraud'].apply(list).reset_index()
job_counts = []
for i in range(job_df.shape[0]):
    job_counts.append(Counter(job_df['is_fraud'][i]))
new_job_df = pd.DataFrame(job_counts)
job_df = pd.concat([job_df,new_job_df],axis=1)
job_df.drop(['is_fraud'],axis=1,inplace=True)
job_df['ratio'] = ((job_df[0] /(job_df[0] + job_df[1]) * 100))
job_df.sort_values(by='ratio',ascending=True)
job_df.fillna(0,inplace=True)
job_df['ratio'] = ((job_df[0] /(job_df[0] + job_df[1]) * 100))
job_df.sort_values(by='ratio',ascending=True)
```

```python
state_df = df.groupby('state')['is_fraud'].apply(list).reset_index()
state_counts = []
for i in range(state_df.shape[0]):
    state_counts.append(Counter(state_df['is_fraud'][i]))
new_state_df = pd.DataFrame(state_counts)
state_df = pd.concat([state_df,new_state_df],axis=1)
state_df.drop(['is_fraud'],axis=1,inplace=True)
state_df['ratio'] = ((job_df[0] /(state_df[0] + state_df[1]) * 100))
state_df.sort_values(by='ratio',ascending=True)
state_df.fillna(0,inplace=True)
state_df['ratio'] = ((state_df[0] /(state_df[0] + state_df[1]) * 100))
state_df.sort_values(by='ratio',ascending=True).head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf1 = ColumnTransformer([
    ('gender_encoded',OneHotEncoder(),[4])
])
trf2 = ColumnTransformer([
    ('label_encoded',LabelEncoder(),[1,2,5,9])
])
trf3 = RandomForestClassifier()
```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3)
])
```

```python
X = df.drop(['is_fraud'],axis=1)
y = df['is_fraud']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=39)
```
