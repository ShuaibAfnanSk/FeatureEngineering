```python
df.dropna(inplace=True)
```

```python
df['Arrival_Time'] = pd.to_datetime(df['Arrival_Time'])
```

```python
df['Arrival_hour'] = df['Arrival_Time'].dt.hour
```

```python
df['Arrival_Minutes'] = df['Arrival_Time'].dt.minute
```

```python
df = df.drop(columns = ['Arrival_Time'],axis=1)
```

```python
df['Dep_Time'] = pd.to_datetime(df['Dep_Time'])
```

```python
df['Dep_hour'] = df['Dep_Time'].dt.hour
```

```python
df['Dep_minutes'] = df['Dep_Time'].dt.minute
```

```python
df = df.drop(columns = ['Dep_Time'],axis=1)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
df['Total_Stops'] = df['Total_Stops'].astype('int64')
```

```python
df['Duration_hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df.drop(6474,axis=0,inplace=True)
```

```python
df['Duration_hour'] = df['Duration_hour'].astype('int64')
```

```python
df.info()
```

```python
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_min'] = df['Duration_min'].replace(np.NaN,'0')
```

```python
df['Duration_min'] = df['Duration_min'].astype('int64')
```

```python
df['Total_Duration'] = df['Duration_hour']*60+df['Duration_min']
```

```python
df = df.drop(columns=['Duration','Duration_hour','Duration_min'],axis=1)
```

```python
df = df.drop(columns=['Route'],axis=1)
```
