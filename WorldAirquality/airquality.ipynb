```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AQI and Lat Long of Countries.csv')
df.head(10)
```

```python
from collections import Counter
```

```python
country_df = df.groupby('Country')['AQI Category'].apply(list).reset_index()
country_counts = []
for i in range(country_df.shape[0]):
    country_counts.append(Counter(country_df['AQI Category'][i]))
new_country_df = pd.DataFrame(country_counts)
country_df = pd.concat([country_df,new_country_df],axis=1)
country_df.drop(['AQI Category'],axis=1,inplace=True)
country_df.head(10)
```

```python
country_df.fillna(0,inplace=True)
country_df
```

```python
subset_country_df = country_df.sort_values(by='Good',ascending=False).head(10)
```

```python
values = subset_country_df['Good'].tolist()
labels = subset_country_df['Country'].tolist()
plt.pie(values,labels=labels,autopct="%1.1f%%")
plt.show()
```

```python
subset_country_df = country_df.sort_values(by='Hazardous',ascending=False).head(10)
```

```python
values = subset_country_df['Hazardous'].tolist()
labels = subset_country_df['Country'].tolist()
plt.pie(values,labels=labels,autopct="%1.1f%%")
plt.show()
```

```python
values = subset_country_df['Hazardous'].tolist()
labels = subset_country_df['Country'].tolist()
sns.barplot(labels,values)
plt.xticks(rotation='vertical')
plt.show()
```
